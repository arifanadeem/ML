```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
df= pd.read_csv('drug200.csv')
df
```

```python
X= df.drop('Drug',axis='columns')
```

```python
X
```

```python
y=df.Drug
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
category_Sex= LabelEncoder()
category_Cholesterol= LabelEncoder()
category_BP= LabelEncoder()
```

```python
X['Sex_n']= category_Sex.fit_transform(X['Sex'])
X['Cholesterol_n']= category_Cholesterol.fit_transform(X['Cholesterol'])
X['BP_n']= category_BP.fit_transform(X['BP'])
```

```python
X
```

```python
X_final= X.drop(['Sex','BP','Cholesterol'],axis='columns')
```

```python
X_final
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X_final, y, test_size=0.10, random_state=42)
```

```python
from sklearn import tree
decision_tree_regression= tree.DecisionTreeClassifier()
```

```python
decision_tree_regression.fit(X_final,y)
```

```python
decision_tree_regression.score(X_final,y)
```

```python
# if we use train test 
decision_tree_regression.fit(X_train,y_train)
```

```python
decision_tree_regression.score(X_test,y_test)
```
