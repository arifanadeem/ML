```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
df= datasets.load_iris()
```

```python
data_set= pd.DataFrame(df['data'],columns=df.feature_names)
```

```python
data_set
```

```python
data_set['target']= df.target
```

```python
data_set
```

```python
df.target_names # name of three flowers-> assign value in targrt 0,1,2
```

```python
data_set['flower_name']= data_set.target.apply(lambda x:df.target_names[x])
```
